```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt


# 1.

temperatura = np.array([random.randint(20, 35) for _ in range(7)])

dias_semana = np.array(["Segunda", "Terça", "Quarta", "Quinta", "Sexta",
"Sábado", "Domingo"])


print(temperatura)
print(dias_semana)


# 2.

dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]
temperaturas = np.random.randint(20, 36, 7)

df = pd.DataFrame({
    "Dia":  dias,
    "Temperatura": temperaturas
})

print(df)

media = df["Temperatura"].mean()
print(f"Média da temperatura da semana: ", media)

maxima = df["Temperatura"].max()
minima = df["Temperatura"].min()

print("Temperatura máxima: ", maxima)
print("Temperatura mínima: ", minima)


# 3.

plt.figure(figsize=(10,5))

plt.plot(dias, temperaturas)
plt.title("Temperatura semanal")
plt.show()


plt.figure(figsize=(10,5))
plt.plot(dias, temperaturas, marker='o')
plt.title("Temperatura semanal (com marker)")
plt.show()


plt.figure(figsize=(10,5))
plt.plot(dias, temperaturas, linestyle='--')
plt.title("Temperatura semanal(com linestyle)")
plt.show()





```
